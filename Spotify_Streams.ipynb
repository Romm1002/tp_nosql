```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pymongo import MongoClient
```

```python
spotify_final = pd.read_csv('./Spotify_final_dataset.csv', index_col="Id")
spotify_final.head(10)
```

```python
spotify_final = spotify_final.groupby('Artist Name')['Total Streams'].size()
spotify_final.to_csv()
```

Ces données datent de plus d'un an donc je cherche des données plus récentes

```python
spotify_2023 = pd.read_csv('./Spotify_2023.csv')
spotify_2023.head(10)
```

```python
spotify_2023.isnull().sum()
```

```python
spotify_2023['key'].fillna(0,inplace=True)
spotify_2023['in_shazam_charts'].fillna(0,inplace=True)

spotify_2023.isnull().sum()
```

```python
spotify_2023['streams'].isna()
```

```python
spotify_2023['streams'] = pd.to_numeric(spotify_2023['streams'], errors='coerce')
songs = spotify_2023[['track_name', 'streams']].sort_values(by='streams', ascending=False)

songs.head(5)
```

```python
spotify_2023.info()
```

```python
artists = spotify_2023['artist(s)_name'].value_counts().head(5)
artists

plt.figure(figsize=(10, 5))


sns.barplot(x=artists.index , y=artists )

plt.ylabel('No. of Tracks',fontsize=14, fontweight='bold')
plt.xlabel('Artist Names',fontsize=14, fontweight='bold')
plt.title('Top 5 Artists with the Most Tracks')
plt.show()
```

```python
artists = spotify_2023.groupby("artist(s)_name")['streams'].sum()
# artists = artists.sort_values(by="streams", ascending=False)

artists = artists.sort_values(ascending=False).head()

artists
```

```python
plt.figure(figsize=(10, 5))

sns.barplot(x=artists.index , y=artists )

plt.ylabel('No. of Streams',fontsize=14, fontweight='bold')
plt.xlabel('Artist Names',fontsize=14, fontweight='bold')
plt.title('Top 5 Artists with the Most Streams')
plt.show()
```
